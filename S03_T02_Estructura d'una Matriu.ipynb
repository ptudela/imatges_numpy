### Crea una matriu bidimensional amb una forma de 5 x 5. Extreu el valor màxim de la matriu, i els valors màxims de cadascun dels seus eixos.

```python
import numpy as np 
a = np.arange(25).reshape(5,5)
print(a)

# Entreu el valor màxim de la matriu
vmax= a.max()
print(vmax)

# els valors màxims de cadascun dels seus eixos
col = a.max(axis=0)
print(col)             # els valors màxims de columnes
row = a.max(axis=1)
print(row)             # els valors màxims de filas
```

### Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : "les matrius es poden transmetre / broadcast si les seves dimensions coincideixen o si una de les matrius té una mida d'1".

```python
# 1. If two arrays have excactly the same shape
import numpy as np 
m1 = np.array([[1,2,3],[4,5,6]])
print(m1)
print(m1.shape)

m2 = np.arange(6).reshape(2,3)
print(m2)
print(m2.shape)

m1*m2    # execute an operation. As both of them have the same shape, the broadcasting can be done.
```

```python
# 2. If two arrays with different dimension but at least one of them are 1.

m3 = np.arange(12).reshape(4,3)    
print(m3)
print(m3.shape)

m4 = np.arange(4).reshape(4,1)
print(m4)
print(m4.shape)

m3 * m4 
```

```python
# If the shape of the two arrays does ot match in any dimension, but if any one array is having shape equal to 1.
m5 = np.arange(4).reshape(4,1)  # 2-D array
print(m5)
print(m5.shape)
m6 = np.array([0,1,2])  # 1-D array
print(m6)
print(m6.shape)
m5 + m6                 #Broadcasting works because at least one of them have shape equal to 1.
```

### Utilitza la Indexació per extreure els valors d'una columna i una fila de la matriu. I suma els seus valors.

```python
x = np.arange(25).reshape(5,5)
print(x)

print(x[0]) # row values

print(x[:, 1])  # column values

row0 = np.sum(x[0])  # row 0 sum of values
print(row0)

col1=np.sum(x[:,1]) # column 1 sum of values
print(col1)
```
